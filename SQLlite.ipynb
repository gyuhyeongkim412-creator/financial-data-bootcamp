````python
def execute_sql_script(conn, sql_file_path):
    """
    SQL 스크립트 파일을 읽어서 실행하는 함수

    Parameters:
        conn: 데이터베이스 연결 객체
        sql_file_path: SQL 파일 경로
    """
    cur = conn.cursor()

    # SQL 파일 읽기
    with open(sql_file_path, 'r', encoding='utf-8') as f:
        sql_script = f.read()

    # 세미콜론으로 명령어 분리
    sql_commands = sql_script.split(';')

    success_count = 0
    error_count = 0

    for command in sql_commands:
        command = command.strip()

        # 빈 명령어나 주석만 있는 경우 건너뛰기
        if not command or command.startswith('--'):
            continue

        try:
            cur.execute(command)
            success_count += 1
        except Exception as e:
            error_count += 1
            print(f"⚠️ 오류 발생: {str(e)[:50]}...")

    conn.commit()
    print(f"\n✅ 실행 완료!")
    print(f"   - 성공: {success_count}개 명령어")
    print(f"   - 오류: {error_count}개 명령어")
````

````python
# SQL 스크립트 실행
execute_sql_script(conn, SQL_FILE_PATH)
````

````python
# ============================================================
# 테이블 생성 SQL
# ============================================================

# 1. 마일리지등급 테이블
cur.execute('''
CREATE TABLE IF NOT EXISTS 마일리지등급 (
    등급명 VARCHAR(10) PRIMARY KEY,
    하한마일리지 INT,
    상한마일리지 BIGINT
)
''')

# 2. 부서 테이블
cur.execute('''
CREATE TABLE IF NOT EXISTS 부서 (
    부서번호 VARCHAR(10) PRIMARY KEY,
    부서명 VARCHAR(50)
)
''')

# 3. 사원 테이블
cur.execute('''
CREATE TABLE IF NOT EXISTS 사원 (
    사원번호 VARCHAR(10) PRIMARY KEY,
    이름 VARCHAR(50),
    영문이름 VARCHAR(100),
    직위 VARCHAR(50),
    성별 VARCHAR(10),
    생일 DATE,
    입사일 DATE,
    주소 VARCHAR(200),
    도시 VARCHAR(50),
    지역 VARCHAR(50),
    집전화 VARCHAR(50),
    상사번호 VARCHAR(10),
    부서번호 VARCHAR(10),
    FOREIGN KEY (부서번호) REFERENCES 부서(부서번호)
)
''')

# 4. 고객 테이블
cur.execute('''
CREATE TABLE IF NOT EXISTS 고객 (
    고객번호 VARCHAR(10) PRIMARY KEY,
    고객회사명 VARCHAR(100),
    담당자명 VARCHAR(50),
    담당자직위 VARCHAR(50),
    주소 VARCHAR(200),
    도시 VARCHAR(50),
    지역 VARCHAR(50),
    전화번호 VARCHAR(50),
    마일리지 INT
)
''')

# 5. 제품 테이블
cur.execute('''
CREATE TABLE IF NOT EXISTS 제품 (
    제품번호 INT PRIMARY KEY,
    제품명 VARCHAR(100),
    포장단위 VARCHAR(100),
    단가 INT,
    재고 INT
)
''')

# 6. 주문 테이블
cur.execute('''
CREATE TABLE IF NOT EXISTS 주문 (
    주문번호 VARCHAR(10) PRIMARY KEY,
    고객번호 VARCHAR(10),
    사원번호 VARCHAR(10),
    주문일 DATE,
    요청일 DATE,
    발송일 DATE,
    FOREIGN KEY (고객번호) REFERENCES 고객(고객번호),
    FOREIGN KEY (사원번호) REFERENCES 사원(사원번호)
)
''')

# 7. 주문세부 테이블
cur.execute('''
CREATE TABLE IF NOT EXISTS 주문세부 (
    주문번호 VARCHAR(10),
    제품번호 INT,
    단가 INT,
    주문수량 INT,
    할인율 REAL,
    PRIMARY KEY (주문번호, 제품번호),
    FOREIGN KEY (주문번호) REFERENCES 주문(주문번호),
    FOREIGN KEY (제품번호) REFERENCES 제품(제품번호)
)
''')

conn.commit()
print("✅ 모든 테이블 생성 완료!")
````
